Moving along and looking for nulls

```python
df.isna().sum()
```

```python
def check_for_nans(df):
    #helper function to check for nans in dataframe
    nan_counts = pd.DataFrame(df.isna().sum(), columns=['nan_count'])
    return nan_counts[nan_counts['nan_count'] > 0]
```

```python
list(check_for_nans(df).index)
```

#### Exploring these nan columns and the columns that are similar to them 

starting with `scheme_management` & `scheme_name`

```python
df.loc[:, ['scheme_management', 'scheme_name']].head()
```

Looks like scheme name can be an unnecessary column as scheme mangement info for some of those Nans is avaiable. Also, close to majority, 21132 entries are Nans and we do not really want to use scheme names to decide if a waterpoint is functional or not. Lets go ahead and drop that column.

```python
df.drop(['scheme_name'], axis=1, inplace=True)
```

explolring `scheme_management`

```python
df.scheme_management.value_counts(dropna=False)
```

Since I don't want to just drop 2922 rows since their `scheme_management` values are missing and also I don't want to just impute the data with a random guess I'll just replace those missing values as `unknown`

```python
def fillnan_as_unknown(df, name):
    #helper function for filling nans with Unknowns 
    df[name].fillna('unknown', inplace=True)
```

```python
fillnan_as_unknown(df, 'scheme_management')
```

```python
check_for_nans(df)
```
